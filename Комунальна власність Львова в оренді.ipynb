# Комунальна власність Львова в оренді

```python
import pandas as pd
```

```python
properties = pd.read_csv("communal-properties.csv")
```

```python
sykhivprops = properties[properties.district == "Сихівський"]
```

```python
sykhivprops.sort_values("square", ascending=False)
```

```python
sykhivprops.holder.value_counts().reset_index()
```

```python
sykhivprops.square.sum()
```

```python
sykhivprops[sykhivprops.holder == "Іваницька Марія Петрівна ФОП"].square.sum()
```

```python
sykhivholders = sykhivprops.holder.value_counts().reset_index()
```

```python
sykhivholders.head(10)
```

```python
uniquesykhivholderstemp = []
```

```python
for index, row in sykhivholders.iterrows():
    uniquesykhivholderstemp.append({'holder': row['index'], 'propscount': row['holder'], 'square': sykhivprops[sykhivprops.holder == row['index']].square.sum()})
    
```

```python
uniquesykhivholders = pd.DataFrame(uniquesykhivholderstemp)
```

```python
uniquesykhivholders.head()
```

```python
uniquesykhivholders.sort_values('square', ascending=False)
```
